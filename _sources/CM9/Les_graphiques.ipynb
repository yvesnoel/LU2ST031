Dans le pas à pas qui suit nous allons utiliser 3 listes comme exemple :

````python
abscisses = [9, 16, 10, 1, 5, 0, 14, 3, 12, 15]
abscisses.sort()
ordonnees1 = [9, 7, 4, 8, 13, 3, 17, 19, 12, 14]
ordonnees2 = [12, 15, 8, 5, 7, 9, 12, 9, 11, 15]
````

## L'import du paquet
Toutt commence par l'importation de la bibliothèque. En fait nous n'allons utiliser qu'un sous ensemble de `Matplotlib` : `pyplot`. De plus, nous allons utiliser un raccourci pour ne pas avoir à écrire `matplotlib.pyplot.plot` mais plus simplement `plt.plot` (la fonction `plot`est donnée ici comme exemple de fonction que nous allons utiliser).
L'import s'écrit donc  :

````python
import matplotlib.pyplot as plt
````

## Le cas le plus simple
Lorsque l'on a besoin d'un seul graphique qui ne fait appel qu'à un seul jeu d'axes (un seul axe des abscisses et un seul axe des ordonnées), il est possible de tracer directement ce dont on a besoin. Pour obtenir des graphiques plus complexes ou pour plus le personnaliser il faudra passer par des figures et plusieurs lignes de commandes mais cela sort de  ce que l'on s'est fixé comme objectif pour ce cours. Nous allons donc nous limiter ici à quelques types de tracés simples et à quelques moyens pour adapter le rendu à ses besoins.

## Le nuage de point avec `scatter`
Pour un nuage de point on peut utiliser la fonction `scatter` (*scatter plot* veut dire nuage de points en anglais).
### La commande de base

````python
plt.scatter(abscisses,ordonnees1)
````

### Les marqueurs
Par défaut, le nuage de points est tracé avec des points bleus. Vous pouvez spécifier le paramètre optionnel `marker` de la fonction `scatter` pour changer le type de marqueurs. Par exemple : `"+"`, `"x"`, `"."`, ... ([plus d'infos ici](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers))

````python
plt.scatter(abscisses,ordonnees1,marker='x')
````

### La taille des marqueurs
Il est parfois utilie de changer la taille des marqueurs en utilisant le paramètre optionnel `s`.

````python
plt.scatter(abscisses,ordonnees1,marker='x',s=100)
````

### La couleur
Pour changer la couleur de tous les marqueurs, on utilise le paramètre optionnel `color`. Les couleurs peuvent peuvent être spécifiées de plusieurs manières mais la plus simple est de donner le nom de la couleur sous la forme d'une chaîne de caractère correspondant au nom de la couleur en anglais. (ces noms peuvent être abrégés en une seule lettres ; `'blue'` : `'b'`, `'red'` : `'r'` ou encore `'black'` : `'k'`)

````python
plt.scatter(abscisses,ordonnees1,marker='x',color='red')
````

Le contour se controle avec `edgecolors` et `linewidth`

Pour changer la couleur des marqueurs individuellement, il faut spécifier une liste avec les couleurs à utiliser pour chaque marqueur. Il existe de nombreuses autres possibilités fort utiles mais qui dépassent les objectifs de ce cours ().

````python
plt.scatter(abscisses,ordonnees1,marker='x',c=['r','r','r','g','g','g','b','b','y','k'],s=100)
````

````python
plt.scatter(abscisses,ordonnees1,s=100,color='r',edgecolors='k',linewidth=3)
````

## Les lignes brisées avec `plot`
Beaucoup d'options vues pour `scatter` sont aussi valables pour plot. Dans les exemples ci-dessous vous découvrirez d'autres.
### Le plus simple

````python
plt.plot(abscisses,ordonnees1)
````

### Modification du style des marqueurs
Les paramètres `marker`, `markerfacecolor` et `markeredgecolor`permettent de modifier le style de marqueurs.

````python
plt.plot(abscisses,ordonnees1,color='r',marker='o',markerfacecolor='y',markeredgecolor='k')
````

### La taile des marqueurs et l'épaisseur du trait
La taille des marqueurs se modifie à l'aide de `markersize`.
L'épaisseur du trait de la ligne brisée se modifie elle avec `linewidth`.

````python
plt.plot(abscisses,ordonnees1,marker='o',markersize=10, linewidth=0.5)
````

### Le type de ligne
Le type de trait est modifié par `linestyle` qui peut prendre notamment les valeurs suivantes : `'-'`, `'--'`, `'-.'`, `':'`, `''`

````python
plt.plot(abscisses,ordonnees1,marker='o',linestyle=':')
````

```{admonition} À vous de jouer
:class: seealso
Essayez de reproduire la figure ci-dessous en utilisant les données `abscisses` et `ordonnees2` déjà définies plus haut.
```

````python
plt.plot(abscisses,ordonnees2,color='k',marker='^',markersize=10,markerfacecolor='y',markeredgecolor='g',linestyle='-.',linewidth=2)
````

Par défaut, les lignes sont affichées et il n'y a pas de marqueurs, mais il est aussi possible de ne garder que les marqueurs en spécifiant `linestyle=''` et retomber sur un affichage équivalent à `scatter` (à part l'individualisation des marqueurs).

````python
plt.plot(abscisses,ordonnees1,marker='o',linestyle='')
````

## Autres types de tracés 
Il existe de nombreux autres types de tracés que nous ne développerons pas ici. On peut tout de même citer :
### `bar`

````python
plt.bar(abscisses,ordonnees1)
````

## Les titres et annotations

### Les titres
Le titre du graphique se définie à l'aide de la fonction `title`. LEs titres des des axes des abscisses et des ordonnées avec les fonctions `xlabel` et `ylabel`

````python
plt.plot(abscisses,ordonnees2,color='r',marker='o',markerfacecolor='y',markeredgecolor='k')
plt.title("Titre de mon graph")
plt.xlabel("Description des abscisses")
plt.ylabel("Description des ordonnées")
````

### Les annotations
Il est possible d'ajouter du texte sur le graphique en utilisant la méthode `text` qui prend comme arguments les coordonées du texte à afficher et bien-sûr la chaîne de caractères voulue. L'alignement ce gère avec `horizontalalignment` et la taille des caractères avec `fontsize`.

````python
plt.plot(abscisses,ordonnees1,color='r',marker='o',markerfacecolor='y',markeredgecolor='k')
plt.text(6,15,"Ce texte est centré sur(6,15)",horizontalalignment='center',fontsize=8)
````

### La légende
La légende s'active grâce à la fonction `legend`. Elle nécessite que le paramètre `label` ait été renseigné pour les différents tracés.

````python
plt.plot(abscisses,ordonnees1,color='r',marker='o',markerfacecolor='y',markeredgecolor='k',label="Etiquette pour la légende")
plt.legend()
````

## Les axes
### Les valeurs limites
Les valeurs limites de chaque axe sont spécifiées avec de 2 valeurs : la limite inférierue et la limite supérieure. Les méthodes sont `xlim` et `ylim`.

````python
plt.plot(abscisses,ordonnees1)
plt.xlim(2,20)
plt.ylim(-20,20)
````

### Les graduations
Les valeurs indiquées sur les axes et les graduations associées s'appellent en anglais des *ticks*. Il est possible de changer l'étendue ou la périodicité de ces graduations grâce à `xticks` (ou `yticks`). Attention les limites du graphes seront éventuellement augmentées pour que toutes les graduations soient visibles. 

````python
plt.plot(abscisses,ordonnees1)
plt.xticks(range(0,30,5))
````

```{admonition} Remarque
:class: note
La méthode `xticks` renvoie deux listes. La première correspond aux graduations créées et la deuxième aux textes correspondants. Ces valeurs renvoyées créent un affichage (souvent non désiré) lorsque l'appel est en dernière ligne d'une cellule et que les valeurs renvoyées ne sont récupérées dans une variable.
```

On peut changer le texte affiché sous chaque graduation mais il faut qu'il en ait exactement le même nombre que les graduations.

````python
plt.plot(abscisses,ordonnees1)
plt.xticks(range(0,21,3),labels=['Lu','Ma','Me','Je','Ve','Sa','Di'])
````

Il est possible de gérer 2 nivaux de graduations. La graduation principale (*major*) vue précédemment et la graduation secondaire (*minor*), généralement sans étiquette. Dans l'exemple suivant, on trace des graduations principale (avec des étiquettes automatiques) de 0 à 16 par pas de 5 et des graduations secondaires (sans étiquette) de 0 à 16 par pas de 1.

````python
plt.plot(abscisses,ordonnees1)
plt.xticks(range(0,17,5))
plt.xticks(range(0,17),minor=True)
plt.show()
````

Ou plus simplement :

````python
plt.plot(abscisses,ordonnees1)
plt.xticks(range(0,17,5))
plt.minorticks_on()
````

Il est possible d'afficher une grille avec la méthode `grid`. [Plus de détails ici](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html)

````python
plt.plot(abscisses,ordonnees1)
plt.xticks(range(0,17,5))
plt.grid()
````

## Plusieurs tracés

### Sur un même graphique
Il est évidemment possible d'afficher plusieurs tracés sur un même graphique.

````python
plt.plot(abscisses,ordonnees1)
plt.scatter(abscisses,ordonnees2,color='r')
````

### Sur des graphiques différents
La méthode `subplot` permet de créer et d'utiliser plusieurs graphique. Les deux premiers chiffres de l'unique argument permetent de spécifier combien de graphiques on va créer (nombre de lignes et de colonnes). Le troisième chiffre indique dans quel graphique on souhaite envoyer les commandes de tracé qui vont suivre. Le numéro 1 est en haut à gauche, on augment en ligne puis en colonne. Dans l'exemple ci-dessous, on crée un mosaïque de 6 graphiques : 2 lignes et 3 colonnes. Les numéros des zones accueillant les graphiques sont donc :
:::Table
| 1 | 2 | 3 |
| 4 | 5 | 6 |
:::


````python
plt.subplot(231)
plt.plot(abscisses,ordonnees1)
plt.subplot(233)
plt.bar(abscisses,ordonnees1)
plt.subplot(235)
plt.scatter(abscisses,ordonnees2,color='r')
````

## Sauvegarde de la figure
Pour sauvegarder la figure que l'on vient de tracer dans un fichier, il suffit de faire appel à la méthode `savefig` et de spécifier le nom de fichier voulu. Le type est automatiquement déduit de l'extension du fichier.

````python
plt.plot(abscisses,ordonnees1)
plt.scatter(abscisses,ordonnees2,color='r')
plt.savefig('./exemple.png')
````
